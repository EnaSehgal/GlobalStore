```python
aggregations = {
    'Quantity': {'Min_Qty': min, 'Max_Qty':max},
    'Discount': {'Sum_Disc': sum,'Mean_Disc':np.mean},
    'Profit':  {'Min_Profit':min,'Max_Profit':max,'Sum_Profit':sum},
    'Sales': {'Min_Sales':min,'Max_Sales':max,
              'Range_Sales':lambda x: max(x) - min(x),
              'Sum_Sales':sum,'Mean_Sales':np.mean}
     }

d18 =  df[(df['Order Priority']=='Medium') & (df['Category']=='Furniture')].groupby(['Sub-Category','Year']).agg(aggregations)
d18
```
